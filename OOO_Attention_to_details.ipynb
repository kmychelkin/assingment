```python
dftest = df1[df1['Выборка'] == "Тестовая"] 
dftest.head()
```

```python
#2 
#задаю ключи для магазинов формата Мини ТЦ
dfmtc = df1[df1['Формат магазина'] == "Мини ТЦ"] 
dfid = dfmtc[['id точки']]
```

```python
#задаю df за 2016 год
df2016 = df2.loc[:, ['id точки','2016-02-01','2016-03-01','2016-04-01','2016-04-01','2016-05-01','2016-06-01','2016-07-01',
                     '2016-08-01','2016-09-01','2016-10-01', '2016-11-01',  '2016-12-01', '2017-01-01']]
df2016.head()
```

```python
#соединияю ключи на двух спредшитах из экселя по выручке 16 года и по формату мини ТЦ
result = pd.merge(dfid, df2016, how='inner', on=['id точки'])
```

```python
#избавляюсь от nan 
result1 = result.dropna()
result1
```

```python
#считаю среднее всех магазинов по 2016 году в столбец mean sales of 2016
#есть в нули в некоторых магазинах, усреднять не стал 
result1['mean sales of 2016'] = result1.iloc[:, 1:].mean(axis = 1)
result1
```

```python
#считаю годовое среднее всех магазинов
res = result1['mean sales of 2016'].mean()
print("среднее всем магазинам в 2016 году:", res)
```
